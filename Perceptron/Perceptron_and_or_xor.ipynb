```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.linear_model import Perceptron
and_data = pd.read_csv(r'D:\DeepLearning\DataSets\and.csv')
or_data = pd.read_csv(r'D:\DeepLearning\DataSets\or.csv')
xor_data = pd.read_csv(r'D:\DeepLearning\DataSets\xor.csv')
```

```python
and_data
```

```python
X_and = and_data.iloc[:,:2]
y_and = and_data.iloc[:,-1]
```

```python
X_and
```

```python
X_or = or_data.iloc[:,:2]
y_or = or_data.iloc[:,-1]
```

```python
X_xor = xor_data.iloc[:,:2]
y_xor = xor_data.iloc[:,-1]
```

```python
print("and table data\n" , X_and , "\n" ,y_and)
```

```python
print("or table data\n" , X_or , "\n" ,y_or)
```

```python
print("xor table data\n" , X_xor , "\n" ,y_xor)
```

```python
model1 = Perceptron()
model2 = Perceptron()
model3 = Perceptron()
```

```python
model1.fit(X_and , y_and)
```

```python
model2.fit(X_or , y_or)
```

```python
model3.fit(X_xor , y_xor)
```

```python
w1 = model1.coef_[0]
b1 = model1.intercept_[0]

print("Weights:", w1)
print("Bias:", b1)
```

```python
w2 = model2.coef_[0]
b2 = model2.intercept_[0]

print("Weights:", w2)
print("Bias:", b2)
```

```python
w3 = model3.coef_[0]
b3 = model3.intercept_[0]

print("Weights:", w3)
print("Bias:", b3)
```

```python
m1 = -(w1[0]/w1[1])
c1 = -(b1/w1[1])
```

```python
x_input = np.linspace(-0.5, 1.5, 100)
y_input = m1*x_input + c1
```

```python
plt.figure(figsize=(6,6))

plt.plot(x_input, y_input, color='red', linewidth=2)

plt.scatter(X_and.iloc[:,0], X_and.iloc[:,1], c=y_and, cmap='winter', s=120)

plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)

plt.grid(True)
plt.show()
```
